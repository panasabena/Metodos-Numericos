## Importamos librer√≠as

```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
## Quiero repasar algunos conceptos de derivada
```

```python
# Definir la funci√≥n f(x, y)
def f(x, y):
    return 3 * x**2 + 5 * x * y - 7 * y**2 + 1

# Definir los rangos de valores para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calcular los valores de f(x, y)
f_values = f(x, y)

# Crear la figura y el subplot 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la funci√≥n f(x, y)
# para agregarle contorno a la gr√°fica hay que cambiar el edge color a 'black'
ax.plot_surface(x, y, f_values, cmap='viridis', edgecolor='none', alpha=0.8)

# Personalizar el gr√°fico
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Funci√≥n 3X^2 + 5xy - 7y^2 + 1')

# Mostrar el gr√°fico
plt.show()
```

#### Ahora agregamos al gr√°fico la funci√≥n derivada con respecto a x

```python
# Definir la funci√≥n f(x, y)
def f(x, y):
    return 3 * x**2 + 5 * x * y - 7 * y**2 + 1

# Definir la derivada parcial df/dx
def df_dx(x, y):
    return 6 * x + 5 * y

# Definir los rangos de valores para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calcular los valores de f(x, y) y df/dx
f_values = f(x, y)
df_dx_values = df_dx(x, y)

# Crear la figura y el subplot 3D
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Graficar la funci√≥n f(x, y) sin contornos
ax.plot_surface(x, y, f_values, cmap='viridis', edgecolor='none')

# Graficar la derivada df/dx como l√≠neas en el plano xy
ax.plot_wireframe(x, y, df_dx_values, colors='grey', linewidths=1)

# Personalizar el gr√°fico
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Funci√≥n sin derivar y derivada en x')

# Mostrar el gr√°fico
plt.show()
```

#### Ahora agregamos la funcion derivada parcial en y 

```python
# Definir las funciones f(x, y), df/dx y df/dy
def f(x, y):
    return 3 * x**2 + 5 * x * y - 7 * y**2 + 1

def df_dx(x, y):
    return 6 * x + 5 * y

def df_dy(x, y):
    return 5 * x - 14 * y

# Definir los rangos de valores para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Calcular los valores de f(x, y), df/dx y df/dy
f_values = f(x, y)
df_dx_values = df_dx(x, y)
df_dy_values = df_dy(x, y)

# Graficar las funciones f(x, y), df/dx y df/dy
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, f_values, cmap='viridis', edgecolor='none', alpha=0.3)
ax.plot_surface(x, y, df_dx_values, cmap='plasma', edgecolor='none', alpha=0.8)
ax.plot_surface(x, y, df_dy_values, cmap='magma', edgecolor='none', alpha=0.8)

# Graficar los contornos de las intersecciones
contour_levels = [0]
ax.contour(x, y, f_values, contour_levels, colors='red', linestyles='dashed')
ax.contour(x, y, df_dx_values, contour_levels, colors='blue', linestyles='dashed')
ax.contour(x, y, df_dy_values, contour_levels, colors='green', linestyles='dashed')

# Personalizar el gr√°fico
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Funciones y contornos de intersecciones')

# Mostrar el gr√°fico
plt.show()
```

### Este c√≥digo calcular√° los valores de f(x, y) y df/dx en una malla de coordenadas (x, y) y luego trazar√° los contornos de intersecci√≥n entre f y df/dx en color rojo y azul respectivament

```python
# Definir la funci√≥n f(x, y)
def f(x, y):
    return 3 * x**2 + 5 * x * y - 7 * y**2 + 1

# Definir la derivada parcial df/dx
def df_dx(x, y):
    return 6 * x + 5 * y

# Definir los rangos de valores para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Crear una malla de coordenadas (x, y)
X, Y = np.meshgrid(x, y)

# Calcular los valores de f(x, y) y df/dx en la malla
Z_f = f(X, Y)
Z_df_dx = df_dx(X, Y)

# Graficar la intersecci√≥n de f y df/dx
plt.contour(X, Y, Z_f, levels=[0], colors='red', linestyles='dashed')
plt.contour(X, Y, Z_df_dx, levels=[0], colors='blue', linestyles='dashed')

# Personalizar el gr√°fico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intersecci√≥n de f y df/dx')

# Mostrar el gr√°fico
plt.show()
```

### Este c√≥digo agregar√° una nueva l√≠nea que grafica la intersecci√≥n entre f y df/dy en color verde. Cada l√≠nea de contorno representa la intersecci√≥n de la funci√≥n f con una de las derivadas parciales df/dx y df/dy en el plano XY.

```python
# Definir la funci√≥n f(x, y)
def f(x, y):
    return 3 * x**2 + 5 * x * y - 7 * y**2 + 1

# Definir las derivadas parciales df/dx y df/dy
def df_dx(x, y):
    return 6 * x + 5 * y

def df_dy(x, y):
    return 5 * x - 14 * y

# Definir los rangos de valores para x y y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# Crear una malla de coordenadas (x, y)
X, Y = np.meshgrid(x, y)

# Calcular los valores de f(x, y), df/dx y df/dy en la malla
Z_f = f(X, Y)
Z_df_dx = df_dx(X, Y)
Z_df_dy = df_dy(X, Y)

# Graficar la funci√≥n f(x, y)
plt.contour(X, Y, Z_f, levels=[0], colors='red', linestyles='dashed')

# Graficar la intersecci√≥n de f y df/dx
plt.contour(X, Y, Z_df_dx, levels=[0], colors='blue', linestyles='dashed')

# Graficar la intersecci√≥n de f y df/dy
plt.contour(X, Y, Z_df_dy, levels=[0], colors='green', linestyles='dashed')

# Personalizar el gr√°fico
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Intersecci√≥n de f, df/dx y df/dy')

# Mostrar el gr√°fico
plt.show()
```
